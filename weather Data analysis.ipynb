Overview:-
    This project involves analyzing historical weather data to extract insights and patterns. The analysis will cover temperature trends, precipitation patterns, and wind speeds using Python and its powerful data analysis libraries.

Objectives:-
    Load and preprocess the weather data.
    Analyze temperature trends over time.
    Examine precipitation patterns.
    Investigate wind speed variations.
    Visualize the findings using plots.
Tools and Libraries:-
    Python
    Pandas
    Matplotlib
    Seaborn

```python
# Importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Weather Data.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.columns
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Convert 'Date/Time' to datetime
df['Date/Time'] = pd.to_datetime(df['Date/Time'])
# Set the date as the index
df.set_index('Date/Time', inplace=True)
```

```python
# Plot temperature trends
plt.figure(figsize=(14, 7))
sns.lineplot(x=df.index, y='Temp_C', data=df)
plt.title('Temperature Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.show()
```

```python
# Plot relative humidity trends
plt.figure(figsize=(14, 7))
sns.lineplot(x=df.index, y='Rel Hum_%', data=df)
plt.title('Relative Humidity Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Relative Humidity (%)')
plt.show()
```

```python
# Plot dew point temperature trends
plt.figure(figsize=(14, 7))
sns.lineplot(x=df.index, y='Dew Point Temp_C', data=df)
plt.title('Dew Point Temperature Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Dew Point Temperature (C)')
plt.show()
```

```python
# Plot wind speed variations
plt.figure(figsize=(14, 7))
sns.lineplot(x=df.index, y='Wind Speed_km/h', data=df)
plt.title('Wind Speed Variations Over Time')
plt.xlabel('Date')
plt.ylabel('Wind Speed (km/h)')
plt.show()
```

```python
# Plot visibility trends
plt.figure(figsize=(14, 7))
sns.lineplot(x=df.index, y='Visibility_km', data=df)
plt.title('Visibility Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Visibility (km)')
plt.show()
```

```python
# Plot pressure variations
plt.figure(figsize=(14, 7))
sns.lineplot(x=df.index, y='Press_kPa', data=df)
plt.title('Pressure Variations Over Time')
plt.xlabel('Date')
plt.ylabel('Pressure (kPa)')
plt.show()
```

```python
# Plot weather conditions frequency
plt.figure(figsize=(14, 7))
sns.countplot(y='Weather', data=df, order=df['Weather'].value_counts().index)
plt.title('Frequency of Different Weather Conditions')
plt.xlabel('Count')
plt.ylabel('Weather')
plt.show()
```

```python
# Extract month and year from the date
df['Month'] = df.index.month
df['Year'] = df.index.year

# Average temperature per month
monthly_avg_temp = df.groupby('Month')['Temp_C'].mean()

plt.figure(figsize=(10, 5))
sns.barplot(x=monthly_avg_temp.index, y=monthly_avg_temp.values, palette='viridis')
plt.title('Average Monthly Temperature')
plt.xlabel('Month')
plt.ylabel('Average Temperature (C)')
plt.show()
```

```python
# Temperature distribution
plt.figure(figsize=(10, 5))
sns.histplot(df['Temp_C'], bins=30, kde=True)
plt.title('Temperature Distribution')
plt.xlabel('Temperature (C)')
plt.ylabel('Frequency')
plt.show()

# Wind Speed distribution
plt.figure(figsize=(10, 5))
sns.histplot(df['Wind Speed_km/h'], bins=30, kde=True)
plt.title('Wind Speed Distribution')
plt.xlabel('Wind Speed (km/h)')
plt.ylabel('Frequency')
plt.show()
```
